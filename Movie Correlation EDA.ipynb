```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r"C:\Users\akars\Downloads\Data Analysis files\movies.csv\movies.csv")
df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(pct_missing*100)))
```

```python
df.dtypes
```

```python
df.boxplot(column = ['gross'])
```

```python
df.drop_duplicates()
```

```python
df.sort_values(by =['gross'], inplace = False, ascending=False)
```

```python
sns.regplot(x = 'budget' , y = 'gross' , data = df , scatter_kws={"color": "orange"}, line_kws={"color": "blue"})
```

```python
sns.regplot(x = 'score' , y = 'gross' , data = df,  scatter_kws={"color": "orange"}, line_kws={"color": "blue"})
```

```python
# Correlation Matrix between all numeric columns

df.corr(method = 'pearson', numeric_only=True)
```

```python
df.corr(method = 'spearman', numeric_only=True)
```

```python
corr_matrix = df.corr(numeric_only=True)

sns.heatmap(corr_matrix,annot=True)

plt.title('Correlation matrix of numeric features')

plt.xlabel('Movie features')

plt.ylabel('Movie features')

plt.show()
```

```python
# Looking at the top 15 compaies by gross revenue

company_gross_sum = df.groupby('company')[['gross']].sum()
company_gross_sum_sorted = company_gross_sum.sort_values(['gross'], ascending = False)[:15]
company_gross_sum_sorted = company_gross_sum_sorted.astype('int64')
company_gross_sum_sorted
```

```python
# listing down annual grossing produced by each company 
annual_gross = df.groupby(['company','year'])[['gross']].sum()[:15]

annual_gross_sorted = annual_gross.sort_values(['gross','company','year'], ascending = False)

annual_gross_sorted
```

```python
company_gross_sum = df.groupby('company')[['gross']].sum()
company_gross_sum_sorted = company_gross_sum.sort_values(['gross'], ascending = False)[:15]
company_gross_sum_sorted = company_gross_sum_sorted.astype('int64')
company_gross_sum_sorted
```

```python
plt.scatter(x = df['budget'], y = df['gross'], alpha = 0.5)
plt.title('Budget vs Gross')
plt.xlabel('budget')
plt.ylabel('gross')
plt.show
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df_numerized.corr(method='pearson')
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
